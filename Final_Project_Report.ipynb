**5.3 Answer:**  

**Residual Patterns:**  

The final Random Forest model achieved a Test MAE of $190,036, closely matching its CV MAE of $191,241(a gap of only ~$1,200), which shows generalization. However, residuals are not uniformly distributed across the full value range. Based on the feature importance distribution (where `finishedsquarefeet12` alone accounts for ~31.5% of predictive weight), the model leans heavily on a small number of structural features, which may cause larger errors for properties where those features are less predictive.

**Overprediction and Underprediction:**  

Tree-based models are known to underpredict at the upper tail of a skewed target distribution, since extreme values are underrepresented in the training data. The target variable `taxvaluedollarcnt` was not log-transformed, meaning the model was exposed to a right-skewed distribution during training. For high-value properties, underprediction is the more likely failure mode.

**Outliers and Unusual Observations:**  

During preprocessing, 149 rows were removed as outliers (values exceeding the 99th percentile across 7 columns). This reduced noise in training but also limited the model's exposure to extreme property values.

**Surprising Findings:**  

A key finding was that feature selection strategy must match the model type. Forward selection is optimized for Linear Regression which severely hurt the tree-based models:

| Model | Full Feature MAE | Forward Selection MAE | Change |
|---|---|---|---|
| Random Forest | $193,562 | $215,068 | +$21,506 |
| Gradient Boosting | $201,991 | $215,926 | +$13,935 |

RF importance-based selection was far more appropriate, reducing the feature set from 24 to 10 with a MAE loss of under $800 for Random Forest. Notably, `log_unitcnt` had near-zero importance (0.0025) despite being log-transformed, confirming that unit count is not a meaningful driver of tax value in this dataset.
